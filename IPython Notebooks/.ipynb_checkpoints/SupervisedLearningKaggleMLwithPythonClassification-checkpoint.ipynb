```python
# Import the Pandas library
import pandas as pd
# Load the train and test datasets to create two DataFrames
train_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
train = pd.read_csv(train_url)
```

```python
#Print the `head` of the train and test dataframes
#train_head=train.head()
train.head()
```

```python
test_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/test.csv"
test = pd.read_csv(test_url)
```

```python
#Print the `head` of the train and test dataframes
#test_head=test.head()
test.head()
```

```python
#remove NaN from Age for both train and test dataset
Age_cols = ['Age']
train[Age_cols] = train[train[Age_cols] > 0][Age_cols]
train_df=train.dropna()
train_df
```

```python
test[Age_cols]= test[test[Age_cols] > 0][Age_cols]
test_df=test.dropna()
test_df
```

```python
#from sklearn import tree
from sklearn import tree
# find the best feature to predict Survival rate
# define X_features and Y_labels
col_names=['Pclass','Age','SibSp','Parch','Fare']
#X_features= train_head[col_names]
X_features= train_df[col_names]
# assign survial to label 
#Y_labels= train_head.Survived
Y_labels= train_df.Survived
#X_features
#Y_labels
# create a decision tree classifier 
clf=tree.DecisionTreeClassifier()
# fit (find patterns in Data) 
clf=clf.fit(X_features, Y_labels)
print(clf.predict(test_df[col_names]))
```
